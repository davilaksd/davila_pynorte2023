### Ciclo Rankine

##### Pressão de entrada na turbina = 4 MPa 
##### Temperatura de entrada na turbina = 400 °C.

#### Determinar: 

(a) Volume específico (m³/kg);

(b) Energia interna (kJ/kg)

(c) Entalpia (kJ/kg);

(d) Entropia(kJ/kg.K).

```python
import CoolProp.CoolProp as CP # Importa o Coolprop

# Se tenho pressão e temperatura no estado 3

P3 = 4 * 1e6 # Converter para Pascal
T3 = 400 + 273.15 # Converter para Kelvin
fluido = 'Water'

d3 = CP.PropsSI('D', 'P', P3, 'T', T3, fluido) # obter densidade
u3 = CP.PropsSI('U', 'P', P3, 'T', T3, fluido) # obter energia interna
h3 = CP.PropsSI('H', 'P', P3, 'T', T3, fluido) # obter entalpia
s3 = CP.PropsSI('S', 'P', P3, 'T', T3, fluido) # obter entropia
fase3 = CP.PhaseSI('P', P3, 'T', T3, fluido) # obter fase

print("(a) Volume específico (m³/kg): ", 1/d3) # 1/d3 para obter o volume específico
print("(b) Energia interna (kJ/kg): ", u3/1000) # Dividir por 1000 para expressar em (kJ/kg)
print("(c) Entalpia (kJ/kg): ", h3/1000) # Dividir por 1000 para expressar em (kJ/kg)
print("(d) Entropia(kJ/kg.K): ", s3/1000) # Dividir por 1000 para expressar em (kJ/kg.K)
```

### Convencionar estados:

### Para um ciclo Rankine:

#### P1 = P4 e P2 = P3
#### s1 = s2 e s3 = s4

```python
# Se tenho pressão e entropia

P4 = 0.01 * 1e6 
s4 = 6.7714 * 1e3
fluido = 'Water'

d4 = CP.PropsSI('D', 'P', P4, 'S', s4, fluido) # obter volume específico
u4 = CP.PropsSI('U', 'P', P4, 'S', s4, fluido) # obter energia interna
h4 = CP.PropsSI('H', 'P', P4, 'S', s4, fluido) # obter entalpia
T4 = CP.PropsSI('T', 'P', P4, 'S', s4, fluido) # obter entropia
q4 = CP.PropsSI('Q', 'P', P4, 'S', s4, fluido) # obter fase

print("Volume específico (m³/kg): ", 1/d4) # 1/d3 para obter o volume específico
print("Energia interna (kJ/kg): ", u4/1000) # Dividir por 1000 para expressar em (kJ/kg)
print("Entalpia (kJ/kg): ", h4/1000) # Dividir por 1000 para expressar em (kJ/kg)
print("Temperatura (ºC): ", T4 - 273.15) # Dividir por 1000 para expressar em (kJ/kg.K)
print("Título: ", q4) # Título
```

```python
# Se tenho pressão e título

P1 = 0.01 * 1e6
q1 = 0
fluido = 'Water'

d1 = CP.PropsSI('D', 'P', P1, 'Q', q1, fluido) # obter densidade
u1 = CP.PropsSI('U', 'P', P1, 'Q', q1, fluido) # obter energia interna
h1 = CP.PropsSI('H', 'P', P1, 'Q', q1, fluido) # obter entalpia
s1 = CP.PropsSI('S', 'P', P1, 'Q', q1, fluido) # obter entropia
T4 = CP.PropsSI('T', 'P', P1, 'Q', q1, fluido) # obter entropia


print("(a) Volume específico (m³/kg): ", 1/d1) # 1/d3 para obter o volume específico
print("(b) Energia interna (kJ/kg): ", u1/1000) # Dividir por 1000 para expressar em (kJ/kg)
print("(c) Entalpia (kJ/kg): ", h1/1000) # Dividir por 1000 para expressar em (kJ/kg)
print("(d) Entropia(kJ/kg.K): ", s1/1000) # Dividir por 1000 para expressar em (kJ/kg.K)
print("(e) Temperatura: ", T4 - 273.15) # Fase
```

```python
# Se tenho pressão e entropia

P2 = 4 * 1e6 
s2 = 0.6492 * 1e3
fluido = 'Water'

d2 = CP.PropsSI('D', 'P', P2, 'S', s2, fluido) # obter volume específico
u2 = CP.PropsSI('U', 'P', P2, 'S', s2, fluido) # obter energia interna
h2 = CP.PropsSI('H', 'P', P2, 'S', s2, fluido) # obter entalpia
T2 = CP.PropsSI('T', 'P', P2, 'S', s2, fluido) # obter entropia

print("Volume específico (m³/kg): ", 1/d2) # 1/d3 para obter o volume específico
print("Energia interna (kJ/kg): ", u2/1000) # Dividir por 1000 para expressar em (kJ/kg)
print("Entalpia (kJ/kg): ", h2/1000) # Dividir por 1000 para expressar em (kJ/kg)
print("Temperatura (ºC): ", T2 - 273.15) # Dividir por 1000 para expressar em (kJ/kg.K)
```

### Calcular calor e trabalho

```python
# Calor da caldeira

q_h = h3 - h2
print("Calor da caldeira (kJ/kg): ", q_h/1000)

# Trabalho da turbina

w_t = h3 - h4
print("Trabalho da turbina (kJ/kg): ", w_t/1000)

# Calor do condensador

q_l = h1 - h4
print("Calor do condensador (kJ/kg): ", q_l/1000)

# Trabalho da bomba

w_b = h1 - h2
print("Trabalho da bomba (kJ/kg): ", w_b/1000)
```

### Calcular o rendimento do ciclo

```python
e = (w_t - w_b)/q_h
print("Rendimento do ciclo: ", round(e * 100, 2), "%")
```
